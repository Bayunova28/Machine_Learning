## K-Nearest Neighbors Classifier 

```python
#split data
X = df_gender.drop('gender', axis = 1)
y = df_gender['gender']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#setup arrays to store training and test accuracy
neighbors = np.arange(1,9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    clf = KNeighborsClassifier(n_neighbors = k)
    clf.fit(X_train, y_train)
    train_accuracy[i] = clf.score(X_train, y_train)
    test_accuracy[i] = clf.score(X_test, y_test) 

print('Train Accuracy: {}\n'.format(train_accuracy))
print('Test Accuracy: {}'.format(test_accuracy))
```

```python
#find good K-value from error rate
error_rate = []

for i in range(1,9):
    clf = KNeighborsClassifier(n_neighbors = i)
    clf.fit(X_train, y_train)
    pred_i = clf.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
    
plt.plot(range(1,9), error_rate, color = 'blue', linestyle ='dashed', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Find Error Rate from K-Value', fontsize = 15, pad = 12, fontweight = 'bold')
plt.xlabel('K-Value', fontsize = 12)
plt.ylabel('Error Rate', fontsize = 12)
plt.show()
```

```python
#visualize knn model for training and testing accuracy
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy', color = 'r')
plt.plot(neighbors, train_accuracy, label = 'Training accuracy', color = 'b')
plt.title('k-NN Classifier Validation Accuracy of Neighbors', fontsize = 15, pad = 12, fontweight = 'bold')
plt.xlabel('Number of Neighbors', fontsize = 12)
plt.ylabel('Accuracy', fontsize = 12)
plt.legend()
plt.show()
```

```python
#build model
clf = KNeighborsClassifier(n_neighbors = 5)
%time clf.fit(X_train, y_train)
print('Training Accuracy :', clf.score(X_train, y_train))  
print('Testing Accuracy :', clf.score(X_test, y_test))  
```

```python
#calculate prediction probability
prob_train = np.squeeze(clf.predict_proba(X_train)[:, 1].reshape(1, -1))
prob_test = np.squeeze(clf.predict_proba(X_test)[:, 1].reshape(1, -1))
print('Probability of Training Set : \n', prob_train)
print('\nProbability of Testing Set : \n', prob_test)
```

```python
#false positive rate, true positive rate, thresholds
fpr1, tpr1, thresholds1 = metrics.roc_curve(y_test, prob_test)
fpr2, tpr2, thresholds2 = metrics.roc_curve(y_train, prob_train)

#auc score
auc1 = metrics.auc(fpr1, tpr1)
auc2 = metrics.auc(fpr2, tpr2)
print('Testing of Area Under the Curve (AUC) :', auc1)
print('Training of Area Under the Curve (AUC) :', auc2)
```

```python
#plot auc 
plt.plot(fpr1, tpr1, color = 'blue', label = 'Testing ROC curve area = %0.2f' % auc1)
plt.plot(fpr2, tpr2, color = 'green', label = 'Training ROC curve area = %0.2f' % auc2)
plt.plot([0,1],[0,1], 'r--')
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.title('Receiver Operating Curve (ROC Curves) of K-Nearest Neighbors Classifier', fontsize = 15, pad = 12, 
          fontweight = 'bold')
plt.xlabel('False Positive Rate', size = 14)
plt.ylabel('True Positive Rate', size = 14)
plt.legend(loc = 'lower right')
plt.show() 
```

```python
#prediction
y_pred = clf.predict(X_test)
print(y_pred)
```

```python
#check MSE & RMSE
mse = metrics.mean_squared_error(y_test, y_pred)
print('Mean Squared Error : '+ str(mse))
rmse = math.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error : '+ str(rmse))
```

```python
#confusion matrix
matrix = pd.DataFrame((metrics.confusion_matrix(y_test, y_pred)), ('Male', 'Female'), ('Male', 'Female'))
print(matrix)

#visualize confusion matrix
heatmap = sns.heatmap(matrix, annot = True, annot_kws = {'size': 12}, fmt = 'd', cmap = 'YlGnBu')
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation = 0, ha = 'right', fontsize = 12)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation = 0, fontsize = 12)

plt.title('Confusion Matrix\n(K-Nearest Neighbors Classifier)', fontsize = 15, pad = 12, fontweight = 'bold')
plt.ylabel('True label', fontsize = 12)
plt.xlabel('Predicted label', fontsize = 12)
plt.show()
```

```python
#classification report
report = metrics.classification_report(y_test, y_pred)
print('Classification Report of K-Nearest Neighbors Classifier : \n')
print(report)
```

## K-Neighbors Regressor 

```python
#split data
X = df_salary.drop('SALARY', axis = 1)
y = df_salary['SALARY']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#setup arrays to store training and test accuracy
neighbors = np.arange(1,9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    reg = KNeighborsRegressor(n_neighbors = k)
    reg.fit(X_train, y_train)
    train_accuracy[i] = reg.score(X_train, y_train)
    test_accuracy[i] = reg.score(X_test, y_test) 

print('Train Accuracy: {}\n'.format(train_accuracy))
print('Test Accuracy: {}'.format(test_accuracy))
```

```python
#find good K-value from error rate
error_rate = []

for i in range(1,9):
    reg = KNeighborsRegressor(n_neighbors = i)
    reg.fit(X_train, y_train)
    pred_i = reg.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
    
plt.plot(range(1,9), error_rate, color = 'blue', linestyle ='dashed', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Find Error Rate from K-Value', fontsize = 15, pad = 12, fontweight = 'bold')
plt.xlabel('K-Value', fontsize = 12)
plt.ylabel('Error Rate', fontsize = 12)
plt.show()
```

```python
#visualize knn model for training and testing accuracy
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy', color = 'r')
plt.plot(neighbors, train_accuracy, label = 'Training accuracy', color = 'b')
plt.title('k-NN Regressor Validation Accuracy of Neighbors', fontsize = 15, pad = 12, fontweight = 'bold')
plt.xlabel('Number of Neighbors', fontsize = 12)
plt.ylabel('Accuracy', fontsize = 12)
plt.legend()
plt.show()
```

```python
#build model
reg = KNeighborsRegressor(n_neighbors = 3)
%time reg.fit(X_train, y_train)
print('Training Accuracy :', reg.score(X_train, y_train))  
print('Testing Accuracy :', reg.score(X_test, y_test))  
```

```python
#prediction
y_pred = reg.predict(X_test)
print(y_pred)
```

```python
#check MAE, MSE & RMSE
mae = metrics.mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error : '+ str(mae))
mse = metrics.mean_squared_error(y_test, y_pred)
print('Mean Squared Error : '+ str(mse))
rmse = math.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error : '+ str(rmse))
```

```python
#visualize model
x = y_test
y = y_pred

plt.title('K-Nearest Regressor Model', fontsize = 15, pad = 12, fontweight = 'bold')
plt.plot(x, y, 'o')

m, b = np.polyfit(x, y, 1)
plt.plot(x, m * x + b, color = 'r')
plt.xlabel('Validation', fontsize = 12)
plt.ylabel('Prediction', fontsize = 12)
plt.show()
```
